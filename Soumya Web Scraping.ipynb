```python
from selenium import webdriver

from selenium.webdriver.common.keys import Keys

from selenium.webdriver.chrome.options import Options

from selenium.webdriver.support.ui import WebDriverWait

from selenium.webdriver.support import expected_conditions as EC 

from selenium.webdriver.common.by import By

import time
```

```python
chromeOptions = Options()

chromeOptions.add_argument("--kiosk")
```

```python
from bs4 import BeautifulSoup
```

```python
driver = webdriver.Chrome(options=chromeOptions)

driver.get("https://www.indiainfoline.com/article/news-sector-others/stock-market-glossary-113111500414_1.html")
```

```python
page_source_overview = driver.page_source
```

```python
soup=BeautifulSoup(page_source_overview,'lxml')
```

```python
soup
```

```python
import pandas as pd
```

```python
Topic_Finders = soup.find_all("div" , class_="current_content serif")
```

```python
Topic_Finders_str=str(Topic_Finders[0])
```

```python
type(Topic_Finders_str)
```

```python
Topic_Finders_str
```

```python
Topic_Finders_str = Topic_Finders_str.replace('<div class="current_content serif"><b>','')
```

```python
Topic_Finders_str = Topic_Finders_str.replace('</div>','')
```

```python
Topic_Finders_str
```

```python
Topic_Finders_str_Split = Topic_Finders_str.split("<br/>\n<br/>\n<b>")
```

```python
len(Topic_Finders_str_Split)
```

```python
Topic_Splitted_List = [topic_list_str.split('</b>') for topic_list_str in Topic_Finders_str_Split]
```

```python
Topic_Splitted_List
```

```python
Glossary_DF = pd.DataFrame(Topic_Splitted_List).iloc[:,0:2]
```
