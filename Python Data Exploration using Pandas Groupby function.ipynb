groupby() function of Pandas 
============================

```python
# Define the grouby against column(s) of the dataframe

iris_grp = iris.groupby('species')

iris_grp
```

```python
# Show all the Groups and the indexes of the rows which are part of the Groups

iris_grp.groups
```

```python
# Show the first observation of each group

iris_grp.first()
```

```python
# Show all the observations in a specific group
# Here we are trying to show all the observations which belongs to 'Setosa' group

iris_grp.get_group('setosa')
```

```python
# We can get the summary values at the group level for a specifc column 
# here we are trying to find the mean Sepal_Length at each group level 

iris_grp['sepal_length'].mean()
```

```python
# here we are trying to find the mean Petal_Length at each group level 

iris_grp['petal_length'].max()
```

```python
# here we are trying to find the mathematical summary for petal_length at each group level 

iris_grp['petal_length'].describe()

# or

iris_grp['petal_length'].describe().unstack()
```

```python
# Get the count of obsevations per group

iris_grp.count()

#or 

iris_grp['sepal_length'].count()
```
