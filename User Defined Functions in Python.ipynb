Now , lets create User Defined Function 

      1) Starting with creation of a very basic user defined function with no Parameters

```python
def Square_Of_4() :
    sqr_of_4 = pow(4,2)
    print(sqr_of_4)
    

Square_Of_4()
```

2) Creating a basic user defined function which takes one parameter abd returns aquare of that number

```python
def Square_Of_any_Num(num) :
    sqr_of_num = pow(num,2)
    print(sqr_of_num)
    

Square_Of_any_Num(5)
```

Note - Difference between 'Parameter' & 'Argumnet '

       Parameter : These are the elements you define while defining a Function 
       Argument : There are the elements you pass to a function while calling a function

3) Returning a value from a User Defined Function

```python
def Square_Of_any_Num(num) :
    sqr_of_num = pow(num,2)
    return sqr_of_num
    

SQ_5 = Square_Of_any_Num(5)

SQ_5
```

```python

def Square_Of_any_Num(num) :
    """ This function returns the Square of 
        any value provided by user as input
        to this function"""
    
    sqr_of_num = pow(num,2)
    return sqr_of_num
    

SQ_5 = Square_Of_any_Num(5)

SQ_5
```

4) Defining user defined function which takes more than one parameter and returns a value

```python
def Raise_To_Any_Power(num,pwr) :
    """ This function returns the power of 
        any value provided by user as input
        to this function"""
    
    pwr_of_num = pow(num,pwr)
    return pwr_of_num
    

Cube_5 = Raise_To_Any_Power(5,3)

Cube_5
```

5) Defining user defined function which takes more than one parameter and returns multiple  value

```python
def Raise_To_Any_Power(num,pwr) :
    """ This function returns the power of 
        any value provided by user as input
        to this function"""
    
    pwr_of_num = pow(num,pwr)
    pwr_of_num_rev = pow(pwr,num)
    return pwr_of_num , pwr_of_num_rev
    

Cube_5 = Raise_To_Any_Power(5,3)

Cube_5

type(Cube_5)   # The function returns a tuple ( it is similar to list but cannot be changed)

Cube_5[0]

Cube_5[1]
```

Some inputs on Tuple data structure in Python
     It is same as List except the fact that Tuple contents are immutable ( cannot be changed)
     
     tuple is defined by enclosing the values within parenthesis  

```python
# Define a tuple 
Tuple_Var = ( 2 , 6 ,9 , 'A')

# View the contents of a tuple
Tuple_Var

# Check the type of a tuple variable
type(Tuple_Var)

# Access individual elements of a tuple ( same as list element access )
Tuple_Var[0]
type(Tuple_Var[0])

Tuple_Var[-1]
type(Tuple_Var[-1])


# unpacking a tuple contents 
var1,var2,var3,var4 = Tuple_Var

var1;var2;var3;var4

type(var1) ; type(var4) 



# Changing the value of tuple contents  
Tuple_Var[-1] = 'Change'               # <-- You will get error as tuple is immutable
```
