```python
from selenium import webdriver

from selenium.webdriver.chrome.options import Options
```

```python
chromeOptions = Options()

chromeOptions.add_argument("--kiosk")
```

```python
from bs4 import BeautifulSoup
```

```python
driver = webdriver.Chrome(options=chromeOptions)

driver.get("https://www.indiainfoline.com/article/news-sector-others/stock-market-glossary-113111500414_1.html")
```

```python
page_source_overview = driver.page_source
```

```python
soup=BeautifulSoup(page_source_overview,'lxml')
```

```python
import pandas as pd
```

```python
Topic_Finders = soup.find_all("div" , class_="current_content serif")
```

```python
Glossary_DF=pd.DataFrame({'Term':[],'Definition':[]})
```

```python

for tag in soup.find_all('b') :
    Glossary_DF = Glossary_DF.append({'Term' : tag.text.replace(':','').strip() , 
                        'Definition' : tag.next_sibling},
                        ignore_index=True )
    
```
