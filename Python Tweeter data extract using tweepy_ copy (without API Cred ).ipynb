```python
Tweet_AOC_DF = pd.DataFrame(
                             { 
                                "Tweet_text" : [] ,  
                                "Tweet_Name" : [] ,
                                "Tweet_Screen_name" : [] ,
                                "Tweet_created" : [] ,
                                "Tweet_Retweeted" : [] ,
                                "Tweet_Favourited" :  [] ,
                                "Tweet_Location" : [] ,
                                "Tweet_Time_zone" : [] ,
                                "Tweet_Geo" : [] ,
                                "Tweet_hashTags" : []
                             })


Tweet_AOC_DF
```

```python


idx = 0

for tweet in Cursor(auth_api.user_timeline, id='AOC').items():
    Tweet_text = tweet.text.encode('utf8')   
    Tweet_Name = tweet.author.name.encode('utf8')
    Tweet_Screen_name = tweet.author.screen_name.encode('utf8')
    Tweet_created = tweet.created_at
    Tweet_Retweeted = tweet.retweeted
    Tweet_Favourited =  tweet.favorited
    Tweet_Location = tweet.user.location.encode('utf8')
    Tweet_Time_zone = tweet.user.time_zone
    Tweet_Geo = tweet.geo
    hash_Tags = [ tags["text"] for tags in tweet.entities.get('hashtags')]
    Tweet_hashTags = ' , '.join(hash_Tags)
    
    Tweet_AOC_DF = Tweet_AOC_DF.append(
                            {
                                "Tweet_text" : Tweet_text ,  
                                "Tweet_Name" : Tweet_Name ,
                                "Tweet_Screen_name" : Tweet_Screen_name ,
                                "Tweet_created" : Tweet_created ,
                                "Tweet_Retweeted" : Tweet_Retweeted ,
                                "Tweet_Favourited" :  Tweet_Favourited ,
                                "Tweet_Location" : Tweet_Location ,
                                "Tweet_Time_zone" : Tweet_Time_zone ,
                                "Tweet_Geo" : Tweet_Geo ,
                                "Tweet_hashTags" : Tweet_hashTags
                            },ignore_index=True )
    
    
    idx += 1
    if idx > 100 :
        break
```
